```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
dic_articulos = []
producto = "xbox"
my_header = {'user-agent':'#################'}
r = requests.get(f'https://listado.mercadolibre.com.co/{producto}#D[A:{producto}]',headers=my_header)


soup = BeautifulSoup(r.text,'html.parser')

articulos = soup.find_all('div', {'class':"ui-search-result__wrapper shops__result-wrapper"})

for i in articulos:
    try:
        titulo = i.find('h2', {'class':'ui-search-item__title shops__item-title'}).text
        precio = i.find('span', {'class':'price-tag-fraction'}).text
        envio = i.find('p', {'class':'ui-search-item__shipping ui-search-item__shipping--free shops__item-shipping-free'}).text
        reviews = i.find('span', {'class':'ui-search-reviews__amount'}).text
        estrellas1 =i.find('svg',{'class':'ui-search-icon ui-search-icon--star ui-search-icon--star-full'})['class']
        
        
        estrellas2 =i.find('svg',{'class':'ui-search-icon ui-search-icon--star ui-search-icon--star-half'})['class']
        x = len(estrellas2)
        x = float(x*0.5)
        y = float(len(estrellas1)+x)
    except AttributeError:
        pass
    temporal = {
        'titulo' : titulo,
        'precio' : precio,
        'reviews' : reviews,
        'estrellas' : y,
        }
    dic_articulos.append(temporal)

```

```python
tabla =pd.DataFrame(dic_articulos)
tabla
```
